```python
import numpy as np
import matplotlib.pyplot as plt

def Uin(t):
    res = -1
    if (int)(2*t)%2 == 0:
        res = 1
    return res

N = 101
time = np.linspace(0,10,N)
U = np.empty(N, float)
for i in range(N):
    U[i]=Uin(time[i])

plt.plot(time,U)
plt.xlabel('t')
plt.ylabel('U_in')
plt.show()
```
